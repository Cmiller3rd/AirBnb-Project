```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../Data_Python_Files/AB_NYC_2019.csv')
air_bnb.info()
```

```python
# Which hosts are the busiest and why?
ab_filtered = air_bnb.drop(labels=['id', 'latitude', 'longitude', 'host_id'], axis=1)
busiest_host = ab_filtered.groupby(['host_name', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365', 'neighbourhood']).count()
busiest_host.sort_values(['calculated_host_listings_count', 'reviews_per_month', 'availability_365'], ascending=[False, False, True])

# The busiest host seems to be Sonder NYC, as they manage 327 separate listings.
# 
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb_groups = air_bnb.groupby('neighbourhood_group').count()
air_bnb_groups

# There are 5 neighbourhood groups available, and Manhattan shows up the most, with over 21000 instances.
```

```python
# Are private rooms the most popular in manhattan?
air_bnb_private = air_bnb.groupby(['neighbourhood_group', 'room_type']).count()
air_bnb_private
# No, entire homes are the most popular in Manhattan
```

```python
# Which hosts are the busiest and based on their reviews?
air_bnb_review_busy = air_bnb.groupby(['reviews_per_month', 'number_of_reviews', 'host_name']).count()
air_bnb_review_busy

# Row NYC and Nalicia seem to be the busiest hosts, 
# given the number of reviews and number of reviews per month for their properties.
```

```python
#Which neighorhood group has the highest average price?
air_bnb_high_price = air_bnb[['neighbourhood_group', 'price']].sort_values('neighbourhood_group', ascending=False)
air_bnb_high_price.groupby('neighbourhood_group').mean().reset_index()

# Manhattan has the highest average price for AirBnB rooms
```

```python
# Which neighbor hood group has the highest total price?
air_bnb_high_single = air_bnb[['neighbourhood_group', 'price']].sort_values('price', ascending=False)
air_bnb_high_single.head(10)

# Queens, Brooklyn, and Manhattan each have facilities that cost $1000 per night.
```

```python
#Which top 5 hosts have the highest total price?
air_bnb_host_price = air_bnb[['host_id', 'host_name', 'price']].sort_values('price', ascending=False)
air_bnb_host_price.head(10)
# The hosts with the highest price are Kathrine, Erin, Jelena, Matt, and Amy
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_bnb_100_rev = air_bnb[air_bnb['number_of_reviews'].ge(100)]
air_bnb_zero_avail_100_rev = air_bnb_100_rev[air_bnb_100_rev['availability_365'] == 0]
air_bnb_zero_avail_100_rev[['host_name', 'host_id', 'number_of_reviews']].sort_values('number_of_reviews', ascending=False)
zero_avail_100_rev_no_dups = air_bnb_zero_avail_100_rev.drop_duplicates('host_name').reset_index(drop=True)
print(f"The following hosts had zero availability and at least 100 reviews:")
for host in list(zero_avail_100_rev_no_dups['host_name']):
    print(host)
```

```python
# What host has the highest total of prices and where are they located?
high_host = air_bnb.groupby(['neighbourhood_group', 'host_name']).sum(numeric_only=True)['price']
high_host.sort_values(ascending=False)

# Sonder NYC in Manhattan has the highest total of prices.
```

```python
# When did Danielle from Queens last receive a review?
neighbourhood_count = air_bnb.groupby('neighbourhood_group').count()
print(neighbourhood_count[['host_id', 'host_name']])
danielle_review = air_bnb[['host_id', 'host_name', 'neighbourhood_group', 'last_review']].sort_values('neighbourhood_group', ascending=False).drop_duplicates('host_id')
queens_danielle = danielle_review[danielle_review['host_name'] == 'Danielle']
queens_danielle[queens_danielle['neighbourhood_group'] == 'Queens']
```
